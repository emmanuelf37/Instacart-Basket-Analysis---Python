```python
#04 Wrangle the data so that it follows consistent logic; for example, rename columns with illogical names and drop columns that don’t add anything to your analysis.
df_customers.head(15)  
```

```python
df_customers_columns_dropped = df_customers.drop(columns = ['date_joined']);
df_customers_columns_dropped.head(15)
```

## Step 5

```python
#05 Complete the fundamental data quality and consistency checks you’ve learned throughout this Achievement; 
#   for example, check for and address missing values and duplicates, and convert any mixed-type data.
df_customers_columns_renamed.describe()
```

```python
for col in df_customers_columns_renamed.columns.tolist():
  weird = (df_customers_columns_renamed[[col]].applymap(type) != df_customers_columns_renamed[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_customers_columns_renamed[weird]) > 0:
    print (col)
```

```python
df_customers_columns_renamed['First Name'] = df_customers_columns_renamed['First Name'].astype('str');
df_customers_data_type_fix = df_customers_columns_renamed
```

```python
df_customers_data_type_fix.isnull().sum()
```

```python
df_customers_dups = df_customers_data_type_fix[df_customers_data_type_fix.duplicated()];
df_customers_dups
```

```python
df_customers_cleaned = df_customers_data_type_fix
```

No Duplicates and No NAN/Missing Values, apparently.
